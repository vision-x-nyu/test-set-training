```python
import pandas as pd
from pathlib import Path
import json
```

```python


results_path = "/home/ellisbrown/workspace/test-set-training/results/experiments/qwen2.5_7b_multi_benchmark_sweep_v3"
results_path = Path(results_path)

# list(results_path.glob("*/*/overall_stats.json"))
results = []
for overall_stats_path in results_path.glob("*/*/overall_stats.json"):
    llm_config_path = overall_stats_path.parent / "llm_config.json"
    metadata_path = overall_stats_path.parent / "metadata.json"

    run_id = overall_stats_path.parent.name

    result = {
        "run_id": run_id,
    }

    with open(llm_config_path, "r") as f:
        llm_config = json.load(f)

    with open(metadata_path, "r") as f:
        metadata = json.load(f)

    with open(overall_stats_path, "r") as f:
        overall_stats = json.load(f)

    result.update(llm_config)
    result.update(metadata)
    result.update(overall_stats)

    results.append(result)

results_df = pd.DataFrame(results)
results_df
```
